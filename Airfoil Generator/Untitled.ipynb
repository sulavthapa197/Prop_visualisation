```python

origin = 50

for i in range(len(propeller)):
    x = []
    y = []
    for j in range(len(y_clark)):
        x_init = (y_clark.iloc[j,0]-(50/100))*propeller.iloc[i,4]
        y_init = y_clark.iloc[j,1]*propeller.iloc[i,4]
        x_final = x_init*np.cos((np.pi/180.)*propeller.iloc[i,2])+y_init*np.sin((np.pi/180.)*propeller.iloc[i,2])
        y_final = y_init*np.cos((np.pi/180.)*propeller.iloc[i,2])-x_init*np.sin((np.pi/180.)*propeller.iloc[i,2])
        x.append(x_final)
        y.append(y_final)
    text_write(x,y,i,propeller.iloc[i,3])
            
```

```python
def text_write(x,y,n,z):
    for i in range(len(x)):
        file = open("demofile"+str(n)+".txt", "a")
        file.write(str(z)+"\t"+str(y[i])+"\t"+str(x[i])+"\n")
        file.close()
```
